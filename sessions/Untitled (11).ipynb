## 2. Long format 

#### The previous COVID-19 DataFrame is presented in the so-called long format, it is also known as tidy format.
#### Long format DataFrames are quite convenient as it allows to keep all relevant information in a single DataFrame. Moreover long formats are computationally convenient (it facilitates vectorized operations).
#### In long format DataFrames: 
1. Each variable forms a column.
2. Each observation forms a row.
3. Each type of observational unit forms a table
