## 2 Wide to Long transformation. MELT Operations
#### Pandas provides the method "melt" to tranform DataFrames from long form to wide equivalent.




### 2.1. Let's melt the cases DataFrame

<img src="https://github.com/thousandoaks/Python4DS102/blob/main/images/meltoperation.png?raw=1">

```python
covingCasesWideFormatDataFrame
```

```python
covingCasesLongFormatDataFrame=covingCasesWideFormatDataFrame.melt(id_vars=["dateRep"],var_name='Country', value_name='cases')
```

```python
covingCasesLongFormatDataFrame
```

### 2.2. Let's melt the deaths DataFrame

<img src="https://github.com/thousandoaks/Python4DS102/blob/main/images/meltoperation2.png?raw=1">

```python
covingDeathsWideFormatDataFrame
```

```python
covingDeathsLongFormatDataFrame=covingDeathsWideFormatDataFrame.melt(id_vars=["dateRep"],var_name='Country', value_name='deaths')
```

```python
covingDeathsLongFormatDataFrame
```

### 2.3. Let's combine both DataFrames into a single one

#### We join the two DataFrames using dateRep and Country as keys

```python
covingCasesAndDeathsLongFormatDataFrame=pd.merge(covingCasesLongFormatDataFrame,covingDeathsLongFormatDataFrame,on=['dateRep','Country'])
```

```python
covingCasesAndDeathsLongFormatDataFrame
```

#### If we need to transform into a 100% Long Form DataFrame:

```python
covingCasesAndDeathsLongFormatDataFrameAlternative=covingCasesAndDeathsLongFormatDataFrame.melt(id_vars=["dateRep",'Country'])
```

```python
covingCasesAndDeathsLongFormatDataFrameAlternative
```
